```python
# lets import the scipy package
import scipy.stats as stats
import math

# lets seed the random values
np.random.seed(10)

# lets take a sample size
sample_size = 1000
sample = np.random.choice(a= data['Total'],
                          size = sample_size)
sample_mean = sample.mean()

# Get the z-critical value*
z_critical = stats.norm.ppf(q = 0.95)  

 # Check the z-critical value  
print("z-critical value: ",z_critical)                                

# Get the population standard deviation
pop_stdev = data['Total'].std()  

# checking the margin of error
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size)) 

# defining our confidence interval
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

# lets print the results
print("Confidence interval:",end=" ")
print(confidence_interval)
print("True mean: {}".format(data['Total'].mean()))
```

```python
# feature engineering on the data to visualize and solve the dataset more accurately

# setting a passing mark for the students to pass on the three subjects individually
passmarks = 40

# creating a new column pass_math, this column will tell us whether the students are pass or fail
data['pass_Percentage'] = np.where(data['Percentage']< passmarks, 'Fail', 'Pass')
data['pass_Percentage'].value_counts().plot.pie(colors = ['lightblue', 'lightgreen'])

plt.title('Pass/Fail in Percentage', fontweight = 30, fontsize = 20)
plt.xlabel('status')
plt.ylabel('count')
plt.savefig("piechart1.pdf", bbox_inches="tight",pad_inches=2, transparent=True)
```

```python
data['pass_Total'] = np.where(data['Total']< 300, 'Fail', 'Pass')
data['pass_Total'].value_counts(dropna = False).plot.pie(colors = ['pink', 'yellow'])

plt.title('Pass/Fail in Total', fontweight = 30, fontsize = 20)
plt.xlabel('status')
plt.ylabel('count')
plt.savefig("piechart2.pdf", bbox_inches="tight",pad_inches=2, transparent=True)
```

```python
# checking which student is fail overall

data['status'] = data.apply(lambda x : 'Fail' if x['AllClear'] == 'Fail' else 'pass', axis = 1)

data['status'].value_counts(dropna = False).plot.pie(colors = ['grey', 'crimson'])
plt.title('overall results', fontweight = 30, fontsize = 20)
plt.xlabel('status')
plt.ylabel('count')
plt.savefig("piechart3.pdf", bbox_inches="tight",pad_inches=2, transparent=True)
```
